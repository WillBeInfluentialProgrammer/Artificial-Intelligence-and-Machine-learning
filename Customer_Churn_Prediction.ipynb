<h4> Customer Churn Prediction <h4>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the dataset
file_path = 'path_to_your_file/Bank Customer Churn Prediction.csv'  # Update this with your file path
data = pd.read_csv('Bank_Customer_Churn_prediction.csv')

# Create the dashboard
fig, axs = plt.subplots(3, 2, figsize=(18, 18))

# Scatter Plot: Age vs. Balance, colored by Churn
scatter = axs[0, 0].scatter(data['age'], data['balance'], c=data['churn'], cmap='viridis', alpha=0.5)
axs[0, 0].set_title('Scatter Plot: Age vs. Balance')
axs[0, 0].set_xlabel('Age')
axs[0, 0].set_ylabel('Balance')
fig.colorbar(scatter, ax=axs[0, 0], label='Churn')
axs[0, 0].grid(True)

# Bar Chart: Number of customers per country
country_counts = data['country'].value_counts()
axs[0, 1].bar(country_counts.index, country_counts, color='skyblue')
axs[0, 1].set_title('Bar Chart: Number of Customers per Country')
axs[0, 1].set_xlabel('Country')
axs[0, 1].set_ylabel('Number of Customers')
axs[0, 1].set_xticks(range(len(country_counts.index)))
axs[0, 1].set_xticklabels(country_counts.index)
axs[0, 1].grid(axis='y')

# Pie Chart: Distribution of customers by gender
gender_counts = data['gender'].value_counts()
axs[1, 0].pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
axs[1, 0].set_title('Pie Chart: Distribution of Customers by Gender')

# Histogram: Distribution of credit scores
axs[1, 1].hist(data['credit_score'], bins=30, color='purple', alpha=0.7)
axs[1, 1].set_title('Histogram: Distribution of Credit Scores')
axs[1, 1].set_xlabel('Credit Score')
axs[1, 1].set_ylabel('Frequency')
axs[1, 1].grid(True)

# Bubble Chart: Age vs. Estimated Salary, with bubble size representing balance
bubble = axs[2, 0].scatter(data['age'], data['estimated_salary'], s=data['balance'] / 1000, c=data['churn'], cmap='viridis', alpha=0.5)
axs[2, 0].set_title('Bubble Chart: Age vs. Estimated Salary')
axs[2, 0].set_xlabel('Age')
axs[2, 0].set_ylabel('Estimated Salary')
fig.colorbar(bubble, ax=axs[2, 0], label='Churn')
axs[2, 0].grid(True)

# Line Graph: Average balance over different age groups
age_groups = data.groupby('age')['balance'].mean()
axs[2, 1].plot(age_groups, marker='o', linestyle='-')
axs[2, 1].set_title('Line Graph: Average Balance over Age Groups')
axs[2, 1].set_xlabel('Age')
axs[2, 1].set_ylabel('Average Balance')
axs[2, 1].grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```
