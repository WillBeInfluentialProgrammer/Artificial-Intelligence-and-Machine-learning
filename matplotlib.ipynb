```python
# Sample data
x = df['Age']
y = df['Fare']

# Assign colors based on conditions
colors = [
    'purple' if age <= 5 else
    'yellow' if 5 < age <= 10 else
    'green' if 20 <= age <= 30 else
    'red' if age > 30 else
    'gray'  # default color for ages not covered by the conditions
    for age in x
]

# Create a scatter plot with specified colors
plt.scatter(x, y, c=colors)
plt.xlabel('Age ')
plt.ylabel('Fare ')
plt.title('Scatter Plot of Age vs Fare')

# Create a scatter plot with specified colors
legend_patches = [
    mpatches.Patch(color='pink', label='1-5 years'),
    mpatches.Patch(color='skyblue', label='5-10 years'),
    mpatches.Patch(color='green', label='20-30 years'),
    mpatches.Patch(color='red', label='>30 years'),
]

plt.legend(handles=legend_patches, title='Age Ranges')
plt.show()
```

```python
df = pd.DataFrame({
    'Age': [2, 7, 25, 40, 50, 3, 8, 22, 28, 35],
    'Fare': [10, 20, 15, 30, 50, 12, 18, 22, 25, 35]
})

# Remove outliers using IQR method
def remove_outliers(data, column):
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    IQR = q3 - q1
    lower_bound = q1 - 1.5 * IQR
    upper_bound = q3 + 1.5 * IQR
    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

# Remove outliers from 'Age' and 'Fare' columns
df = remove_outliers(df, 'Age')
df = remove_outliers(df, 'Fare')

# Sample data
x = df['Age']
y = df['Fare']

# Assign colors based on conditions
colors = [
    'purple' if age <= 5 else
    'yellow' if 5 < age <= 10 else
    'green' if 20 <= age <= 30 else
    'red' if age > 30 else
    'gray'  # default color for ages not covered by the conditions
    for age in x
]

# Create a scatter plot with specified colors
plt.scatter(x, y, c=colors)
plt.xlabel('Age ')
plt.ylabel('Fare ')
plt.title('Scatter Plot of Age vs Fare')

# Create a scatter plot with specified colors
legend_patches = [
    mpatches.Patch(color='pink', label='1-5 years'),
    mpatches.Patch(color='skyblue', label='5-10 years'),
    mpatches.Patch(color='green', label='20-30 years'),
    mpatches.Patch(color='red', label='>30 years'),
]

plt.legend(handles=legend_patches, title='Age Ranges')
plt.show()
```
